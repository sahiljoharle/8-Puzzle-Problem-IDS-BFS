# 8-Puzzle Programming

```python

from collections import deque
import time
```

```python
class Info:
    def __init__(self, state, parent, shift, depth, cost, key):
        self.state = state
        self.parent = parent # Parent Node
        self.shift = shift #move up , down, left, right
        self.depth = depth
        self.cost = cost
        self.key = key
        if self.state:
            self.map = ''.join(str(e) for e in self.state)
    def __el__(self, other):
        return self.map == other.map
    def __id__(self, other):
        return self.map < other.map
    def __str__(self):
        return str(self.map)
def print_out(state):
    for a, b, c in zip(*[iter(state)]*3):
        print(a,b,c)
        
```

```python
goal_State = [0, 1, 2, 3, 4, 5, 6, 7, 8]
N_Exp = 0
goal_Node = None 
MaximunSearch = 0
Fq = 0
```

```python
def Create_Children(node):

    global N_Exp
    N_Exp = N_Exp+1

    next_Path = []
    next_Path.append(Info(Direction_to_shift(node.state, 1), node, 1, node.depth + 1, node.cost + 1, 0))
    next_Path.append(Info(Direction_to_shift(node.state, 2), node, 2, node.depth + 1, node.cost + 1, 0))
    next_Path.append(Info(Direction_to_shift(node.state, 3), node, 3, node.depth + 1, node.cost + 1, 0))
    next_Path.append(Info(Direction_to_shift(node.state, 4), node, 4, node.depth + 1, node.cost + 1, 0))
    node1=[]
    for procedure_Path in next_Path:
        if(procedure_Path.state!=None):
            node1.append(procedure_Path)
    return node1
```

```python
def Direction_to_shift(state, directions):
    
    new_State = state[:]
    
    
    index = new_State.index(0)

    if(index==0):
        if(directions==1):
            return None
        if(directions==2):
            temp=new_State[0]
            new_State[0]=new_State[3]
            new_State[3]=temp
        if(directions==3):
            return None
        if(directions==4):
            temp=new_State[0]
            new_State[0]=new_State[1]
            new_State[1]=temp
        return new_State      
    if(index==1):
        if(directions==1):
            return None
        if(directions==2):
            temp=new_State[1]
            new_State[1]=new_State[4]
            new_State[4]=temp
        if(directions==3):
            temp=new_State[1]
            new_State[1]=new_State[0]
            new_State[0]=temp
        if(directions==4):
            temp=new_State[1]
            new_State[1]=new_State[2]
            new_State[2]=temp
        return new_State    
    if(index==2):
        if(directions==1):
            return None
        if(directions==2):
            temp=new_State[2]
            new_State[2]=new_State[5]
            new_State[5]=temp
        if(directions==3):
            temp=new_State[2]
            new_State[2]=new_State[1]
            new_State[1]=temp
        if(directions==4):
            return None
        return new_State
    if(index==3):
        if(directions==1):
            temp=new_State[3]
            new_State[3]=new_State[0]
            new_State[0]=temp
        if(directions==2):
            temp=new_State[3]
            new_State[3]=new_State[6]
            new_State[6]=temp
        if(directions==3):
            return None
        if(directions==4):
            temp=new_State[3]
            new_State[3]=new_State[4]
            new_State[4]=temp
        return new_State
    if(index==4):
        if(directions==1):
            temp=new_State[4]
            new_State[4]=new_State[1]
            new_State[1]=temp
        if(directions==2):
            temp=new_State[4]
            new_State[4]=new_State[7]
            new_State[7]=temp
        if(directions==3):
            temp=new_State[4]
            new_State[4]=new_State[3]
            new_State[3]=temp
        if(directions==4):
            temp=new_State[4]
            new_State[4]=new_State[5]
            new_State[5]=temp
        return new_State
    if(index==5):
        if(directions==1):
            temp=new_State[5]
            new_State[5]=new_State[2]
            new_State[2]=temp
        if(directions==2):
            temp=new_State[5]
            new_State[5]=new_State[8]
            new_State[8]=temp
        if(directions==3):
            temp=new_State[5]
            new_State[5]=new_State[4]
            new_State[4]=temp
        if(directions==4):
            return None
        return new_State
    if(index==6):
        if(directions==1):
            temp=new_State[6]
            new_State[6]=new_State[3]
            new_State[3]=temp
        if(directions==2):
            return None
        if(directions==3):
            return None
        if(directions==4):
            temp=new_State[6]
            new_State[6]=new_State[7]
            new_State[7]=temp
        return new_State
    if(index==7):
        if(directions==1):
            temp=new_State[7]
            new_State[7]=new_State[4]
            new_State[4]=temp
        if(directions==2):
            return None
        if(directions==3):
            temp=new_State[7]
            new_State[7]=new_State[6]
            new_State[6]=temp
        if(directions==4):
            temp=new_State[7]
            new_State[7]=new_State[8]
            new_State[8]=temp
        return new_State
    if(index==8):
        if(directions==1):
            temp=new_State[8]
            new_State[8]=new_State[5]
            new_State[5]=temp
        if(directions==2):
            return None
        if(directions==3):
            temp=new_State[8]
            new_State[8]=new_State[7]
            new_State[7]=temp
        if(directions==4):
            return None
        return new_State
```

```python
def Breadth_First_Search(state):
    global Fq, goal_Node, MaximunSearch

    visited= set()
    queue = deque([Info(state, None, None, 0, 0, 0)])

    while queue:
        node = queue.popleft()
        visited.add(node.map)
        if node.state == goal_State:
            goal_Node = node
            return queue
        posible_Path = Create_Children(node)
        for path in posible_Path:
            if path.map not in visited:
                queue.append(path)
                visited.add(path.map)
                if path.depth > MaximunSearch:
                    MaximunSearch = MaximunSearch + 1
        if len(queue) > Fq:
            queue_Size = len(queue)
            Fq = queue_Size
            
```

```python
def iddfs(state, depth = 20):
    global Fq, goal_Node, MaximunSearch
    depth_limit = depth
    visited= set()
    queue = deque([Info(state, None, None, 0, 0, 0)])
    global count = 0
    while queue:
        node = queue.popleft()
        count =count +1
        visited.add(node.map)
        if node.state == goal_State:
            print("The number of nodes visited",count)
            return queue
        if depth < depth_limit:
            posible_Path = Create_Children(node)
            for path in posible_Path:
                if path.map not in visited:
                    queue.append(0,path)
            
def ids(Original_State,goal_Node, depth = 50 ):
    for i in range(depth):
        result = iddfs(Original_State,i)
        if result != None:
            return result
        
        
```

```python
def print_result():
    file1 = open('Input8PuzzleCases.txt', 'r')
    global Fq, goal_Node, MaximunSearch,count
    total_step =0
    start=time.time()
    ct=0
    for line in file1:
        ct=ct+1
        line=line.split(", ")
        Original_State=[]
        for l in line:
            Original_State.append(int(l))
        Breadth_First_Search(Original_State)
        result = ids( Original_State,goal_Node) 
        shift = []

        while Original_State != goal_Node.state:
            shift.insert(0, goal_Node.state)
            goal_Node = goal_Node.parent

        shift.insert(0,Original_State)

        shift.pop()
        for value in shift:
            print_out(value)
            print("\nTo\n")
        print_out(goal_State)
        total_step = total_step + len(shift)

        print("\nTime=",time.time()-start)
        print("Steps=", len(shift),"\n---------------\n\n")



    end=time.time()
    timer=end-start
    avg_time=timer/ct
    print("Avgerage Time=",avg_time)
    print("Average steps=",total_step/100)
```
